(b)Using the `rnorm()` function, create a vector, `eps`, containing `100` observations drawn from a `N(0,0.25)` distribution—a normal distribution with mean zero and variance `0.25`.

```r
eps = rnorm(100, 0, sqrt(0.25))
sd(eps)
```
